```python
import json
import os
import uuid
import jwt
import datetime
from datetime import datetime, timedelta
import aiohttp
import asyncio
import boto3
from boto3.dynamodb.conditions import Attr
from botocore.exceptions import ClientError
import yandexcloud
from yandex.cloud.lockbox.v1.payload_service_pb2 import GetPayloadRequest
from yandex.cloud.lockbox.v1.payload_service_pb2_grpc import PayloadServiceStub
from boto3.dynamodb.conditions import Key
import bcrypt
import logging
logger = logging.getLogger('differential_evolution_logger')
logger.setLevel(logging.DEBUG)
boto_session = None
def get_boto_session():
    global boto_session
    if boto_session is not None:
        return boto_session

    yc_sdk = yandexcloud.SDK()
    channel = yc_sdk._channels.channel("lockbox-payload")
    lockbox = PayloadServiceStub(channel)
    response = lockbox.Get(GetPayloadRequest(secret_id=os.environ['SECRET_ID']))

    access_key = None
    secret_key = None
    for entry in response.entries:
        if entry.key == 'ACCESS_KEY_ID':
            access_key = entry.text_value
        elif entry.key == 'SECRET_ACCESS_KEY':
            secret_key = entry.text_value
    if access_key is None or secret_key is None:
        raise Exception("secrets required")
    print("Key id: " + access_key)

    # initialize boto session
    boto_session = boto3.session.Session(
        aws_access_key_id=access_key,
        aws_secret_access_key=secret_key
    )
    return boto_session

boto_session = get_boto_session()
dynamodb = boto_session.resource('dynamodb', endpoint_url=os.environ['DOCAPI_ENDPOINT'], region_name='ru-central1')
users_table = dynamodb.Table('users')
tokens_table = dynamodb.Table('tokens')

def return_response (params):
  return {
        'statusCode': 200,
        'headers': {
            'Content-Type': 'application/json'
        },
        'body': json.dumps(params),
        'isBase64Encoded': False
        }
def create_user(login, password):
    user_id = str(uuid.uuid4())
    response = users_table.scan(
        FilterExpression=Attr('login').eq(login)
    )
    not_ex = True if response['Items'] else False
    if not_ex:
        return return_response ({
            'error': 'already existing acc'
        })
    try:
        users_table.put_item(Item={
            'user-id': user_id,
            'login': login,
            'password': str(password)
        })
        return return_response ({
            'user-id': user_id
        })
    except ClientError as e:
        return return_response ({
            'error': str(e)
        })

SECRET_KEY = os.environ['JWT_KEY']

def user_login(login, mail, telegram, password):

    response = users_table.scan(
        FilterExpression=Attr('login').eq(login) | Attr('mail').eq(mail) | Attr('telegram').eq(telegram)
    )
    user = response['Items'][0] if response['Items'] else None
    if not user:
        return return_response ({'error': 'not found acc'})

    if user['password'] != password:
        return return_response ({'error': 'invalid password'})

    exp_date = datetime.utcnow() + timedelta(days=90)
    token = jwt.encode({'user-id': user['user-id'], 'exp': exp_date}, SECRET_KEY, algorithm='HS256')

    tokens_table.put_item(Item={
        'token': token,
        'exp-date': exp_date.strftime('%Y-%m-%d %H:%M:%S')
    })

    return return_response ({'jwt_token': token})

def update_user(field, user_id, value, token):
    if not token:
        return return_response ({'error': 'not authorized'})

    try:
        token_data = jwt.decode(token, SECRET_KEY, algorithms=['HS256'])
    except jwt.ExpiredSignatureError:
        return return_response ({'error': 'token expired'})
    except jwt.InvalidTokenError:
        return return_response ({'error': 'invalid token'})

    response = tokens_table.get_item(Key={'token': token})
    if 'Item' not in response or response['Item']['exp-date'] < datetime.now().strftime('%Y-%m-%d %H:%M:%S'):
        return return_response ({'error': 'token expired'})

    try:
        users_table.update_item(
            Key={'user-id': user_id},
            UpdateExpression=f'SET {field} = :val',
            ExpressionAttributeValues={':val': value}
        )
        return return_response ({'message': 'update successful'})
    except ClientError as e:
        return return_response ({
            'error': str(e)
        })

def get_user_info(user_id, token):
    if not token:
        return return_response ({'error': 'not authorized'})

    try:
        token_data = jwt.decode(token, SECRET_KEY, algorithms=['HS256'])
    except jwt.ExpiredSignatureError:
        return return_response ({'error': 'token expired'})
    except jwt.InvalidTokenError:
        return return_response ({'error': 'invalid token'})

    response = users_table.scan(
        FilterExpression=Attr('user-id').eq(user_id)
    )
    if not response['Items']:
        return return_response ({'error': 'User not found'})

    try:
        user = users_table.get_item(Key={'user-id': user_id})['Item']
        user.pop("password")
        return return_response(user)
    except ClientError as e:
        return return_response ({
            'error': str(e)
        })

def handle_api(event, context):
    logger.debug(
            f"{event}")
    event = event['queryStringParameters']
    action = event['action']
    if action == 'create':
        return create_user(event['login'], event['password'])
    elif action == 'login':
        return user_login(event['login'], event['mail'], event['telegram'], event['password'])
    elif action == 'update':
        return update_user(event['field'], event['user-id'], event['value'], event['token'])
    elif action == 'get-user-info':
        return get_user_info(event['user-id'], event['token'])
    else:
        return return_response ({
            'error': str({"error": "unknown action: " + action})
        })
```
